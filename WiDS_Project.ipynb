### Categorical metadata

```python
df_categorical_metadata = pd.read_excel('TRAIN/TRAIN_CATEGORICAL_METADATA.xlsx')
df_categorical_metadata.head()
```

```python
df_categorical_metadata = pd.read_excel('TRAIN/TRAIN_CATEGORICAL_METADATA.xlsx')
df_categorical_metadata.isna().sum()
```

```python
import matplotlib.pyplot as plt

# Define columns and labels
barratt_columns = [
    'Barratt_Barratt_P1_Edu', 'Barratt_Barratt_P1_Occ', 
    'Barratt_Barratt_P2_Edu', 'Barratt_Barratt_P2_Occ'
]
barratt_labels = [
    'Parent 1 Education', 'Parent 1 Occupation', 
    'Parent 2 Education', 'Parent 2 Occupation'
]

# Figure size
plt.figure(figsize=(12, 6))

# Bar width and participant indices
bar_width = 0.2
participants = df_categorical_metadata.index

# Plot bars for each category
for i, (col, label) in enumerate(zip(barratt_columns, barratt_labels)):
    if col in df_categorical_metadata.columns:
        plt.bar(participants + i * bar_width, df_categorical_metadata[col], width=bar_width, label=label, alpha=0.7)
    else:
        print(f"Warning: Column '{col}' not found in DataFrame!")

# Title and axis labels
plt.title('Barratt Features Across Participants')
plt.xlabel('Participant Index')
plt.ylabel('Barratt Score')
plt.xticks(participants + bar_width, participants, rotation=90)  # Adjust x-axis labels
plt.legend()
plt.grid(True, axis='y', alpha=0.3)

# Show the plot
plt.show()
```

```python
# Lineplot for Barratt_Barratt_P1_Edu vs Barratt_Barratt_P1_Occ
plt.figure(figsize=(10,5))
sns.lineplot(
    x='Barratt_Barratt_P1_Edu', 
    y='Barratt_Barratt_P1_Occ', 
    data=df_categorical_metadata, 
    marker='o'
)
plt.title('Lineplot: Barratt_Barratt_P1_Edu vs Barratt_Barratt_P1_Occ')
plt.xlabel('Barratt_Barratt_P1_Edu')
plt.ylabel('Barratt_Barratt_P1_Occ')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Lineplot for Barratt_Barratt_P2_Edu vs Barratt_Barratt_P2_Occ
plt.figure(figsize=(10,5))
sns.lineplot(
    x='Barratt_Barratt_P2_Edu', 
    y='Barratt_Barratt_P2_Occ', 
    data=df_categorical_metadata, 
    marker='o'
)
plt.title('Lineplot: Barratt_Barratt_P2_Edu vs Barratt_Barratt_P2_Occ')
plt.xlabel('Barratt_Barratt_P2_Edu')
plt.ylabel('Barratt_Barratt_P2_Occ')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Lineplot for Barratt_Barratt_P1_Edu vs Barratt_Barratt_P2_Edu
plt.figure(figsize=(10,5))
sns.lineplot(
    x='Barratt_Barratt_P1_Edu', 
    y='Barratt_Barratt_P2_Edu', 
    data=df_categorical_metadata, 
    marker='o'
)
plt.title('Lineplot: Barratt_Barratt_P1_Edu vs Barratt_Barratt_P2_Edu')
plt.xlabel('Barratt_Barratt_P1_Edu')
plt.ylabel('Barratt_Barratt_P2_Edu')
plt.grid(True, alpha=0.3)
plt.show()
```
